Question no 1 : Are movies that are more popular (operationalized as having more ratings) rated higher than movies that
are less popular?

```python
# review count per movie

df_modified = df_1.count().reset_index()

df_modified.rename(columns = {'index' : 'movie_names' ,0 : 'review_count'} , inplace = True)
```

```python
df_modified.head()
```

```python
# finding the median for review count
median_review_count = np.median(df_modified['review_count'].sort_values(ascending = True))
```

```python
median_review_count
```

```python
# All movies above median are popular and below median are not popular

popular_movies = df_modified[df_modified['review_count']>=median_review_count]['movie_names'].to_list()
not_popular_movies = df_modified[df_modified['review_count']<median_review_count]['movie_names'].to_list()
```

```python
# splitting the dataset into popular and non popular

popular_movies_df = df_1[popular_movies]
non_popular_movies_df = df_1[not_popular_movies]
```

```python
popular_movies_df
```

```python
def wrangle_movie_ratings(a):

  """ convert ratings into a single array """
  a1 = a.copy()
  movie_ratings = a1.values.flatten()
  movie_ratings = movie_ratings[~np.isnan(movie_ratings)]
  return movie_ratings
```

```python
popular_movie_ratings = wrangle_movie_ratings(popular_movies_df)
non_popular_movie_ratings = wrangle_movie_ratings(non_popular_movies_df)
```

```python
len(non_popular_movie_ratings)
```

```python
# performing the A Mann-Whitney test

stat , p_value = mannwhitneyu(popular_movie_ratings, non_popular_movie_ratings, alternative='greater')
```

```python
print("p-value is :" , print(mp.mpf(p_value)))
```

```python
print(stat)
```

```python
from decimal import Decimal

small_number = np.double(p_value)  # Replace with your variable
print(f"{small_number:.3500f}")
```

```python
# checking the significance
if p_value<0.005:
  print(f"There is significant difference in ratings : Popular movies are rated higher than Non-popular movies")
else:
  print(f"There is no significant difference in ratings : Popular movies are not rated higher than Non-popular movies")
```
