### Problem 3: We will test if ratings of shrek are gendered with a Mann-Whitney U Test

```python
#So we need to create a shrek and gender database
Quest3 = InitData[["Shrek (2001)", "Gender identity (1 = female; 2 = male; 3 = self-described)"]]
#and clean it
Cle3 = Quest3.dropna()
```

```python
#Next break it up into men and women
fem3 = Cle3[Cle3["Gender identity (1 = female; 2 = male; 3 = self-described)"] == 1]
man3 = Cle3[Cle3["Gender identity (1 = female; 2 = male; 3 = self-described)"] == 2]
```

```python
#Then run the Mann Whitnety Test
stats.mannwhitneyu(fem3["Shrek (2001)"], man3["Shrek (2001)"])
```

And we see this is not significant at p = 0.05

### Problem 4: We will create a function that performs the Mann-Whitney U test from 3 and loop it across all movies

```python
#Our Function
def quest4(movie):
    Quest4 = InitData[[movie, "Gender identity (1 = female; 2 = male; 3 = self-described)"]]
    Cle4 = Quest4.dropna()
    fem4 = Cle4[Cle4["Gender identity (1 = female; 2 = male; 3 = self-described)"] == 1]
    man4 = Cle4[Cle4["Gender identity (1 = female; 2 = male; 3 = self-described)"] == 2]
    results = stats.mannwhitneyu(fem4[movie], man4[movie])[1]
    return results
```

```python
#Our Loop
movies = []
pvalue = []

#I need to grab the column names for all the movies (put them in a listO
for x in range(0,400):
    title = str(InitData.columns[x])
    movies.append(title)
    pvalue.append(quest4(title))

Outputframe = pd.DataFrame({'movies' : movies,'pvalue' : pvalue}, columns = ['movies','pvalue'])
```

```python
#Lets find our ratio
print(len(Outputframe[Outputframe['pvalue'] < 0.005])/400)
```

```python
#I'm also exporting the data to look at the values
Sortframe = Outputframe.sort_values(by = 'pvalue')
Sortframe = Sortframe.reset_index(drop = True)
Sortframe.to_csv("Question 4 Sorted.csv")
```

We find that 12.5% of movies are rated differently between men and women. The movies that make this up like "The Proposal" and "Batman: The Dark Knight" make sense

### Problem 9: We will check the distributions of Home Alone and Finding Nemo with a KS test

```python
#this will be an element wise drop so I will create two new dataframes
Home = InitData['Home Alone (1990)']
Nemo = InitData['Finding Nemo (2003)']

Home = Home.dropna()
Nemo = Nemo.dropna()
```

### Problem 10: We will check the consistency of several movie franchises with a Kruskall Wallis test on each

```python
#Make a function to run on each movie franchise
def series(word):
    #create frame
    Frame = InitData.filter(like = word)
    Movies = []
    for x in Frame.columns:
        Movies.append(Frame[x].dropna())
    return stats.kruskal(*Movies)
```

```python
series('Star Wars')
```
